```python
import numpy as np
import seaborn as sns; sns.set_theme()
x1 = np.arange(100).reshape(-1,1)
x2 = np.arange(100).reshape(1,-1)
uniform_data = np.random.rand(20, 20)
ax = sns.heatmap(uniform_data)
```

```python
X = x1 - x2
ax = sns.heatmap(X + 10 * np.random.rand(100, 100));
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
```

```python
X = x1 - x2
X = 3 * np.exp(-X**2 / 40)
ax = sns.heatmap(X + np.random.rand(100, 100));
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
```

```python
X = x1 - x2
X = 3 * np.cos(X * np.pi / 40)
ax = sns.heatmap(X + np.random.rand(100, 100));
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
```

```python
X = x1 - x2
X = 3 * np.cos(X * np.pi / 10) * np.exp(-X**2 / 1000)
ax = sns.heatmap(X + np.random.rand(100, 100));
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)
```
