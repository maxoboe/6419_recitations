```python
import numpy as np
import matplotlib.pyplot as plt

# thing = np.array([1, -4])
thing = np.array([1, 0])
# thing = np.array([1, -4])
cov = np.array([thing, np.flip(thing)])
print(cov)
X = np.random.multivariate_normal((0,0), cov, size=1000)
plt.hist(X[:,0])
plt.xlabel('X1')
plt.ylabel('count');
plt.gca().axis('equal');
plt.gca().grid(True);
```
