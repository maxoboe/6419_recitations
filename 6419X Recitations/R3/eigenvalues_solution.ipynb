# Eigenvalues and Eigenvectors
In this example, you will simulate 2-dimensional data and use it to visualize the role of eigenvalues in summarizing a covariance matrix. 



```python
# First, load necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

## Part 1: Simulate Data
Generate 100 samples of normally distributed, 2-D data with mean vector $[2, 3]$ and covariance matrix $\displaystyle \Sigma = \begin{bmatrix}
1 & 0.8 \\
0.8 & 2
\end{bmatrix}$.
Visualize the data. 

```python
# The covariance matrix and mean array are given; please fill in:
# 1. Code to generate 100 samples of data, labeled X
# 2. Code to plot X (that is, plot the first component vs the second component)
covariance = np.array([[1, 0.8],[0.8, 2]])
mean = np.array([2,3])
X = np.random.multivariate_normal(mean, covariance,size=100)
plt.scatter(X[:,0],X[:,1])
```

```python
X[0]
```

## Part 2
Find the eigenvalues and eigenvectors of your covariance matrix $\Sigma$. 

```python
# provide code to fill in the eigenvalues and vectors
vals, vectors = np.linalg.eig(covariance)
print(vals,vectors)
```

## Part 3
Center your data (that is, remove the mean from each observation) and compute $(X - \mu)^T(X - \mu)$. 

Find the eigenvalues of this matrix. 

```python
# Provide code to fill in empirical_covariance 
empirical_covariance = (X - mean).T@(X-mean)
empirical_vals, empirical_vectors = np.linalg.eig(empirical_covariance)
print(empirical_vals, empirical_vectors)
```

## Part 4
Add the empirical eigenvalues and true eigenvalues to your scatter plot from earlier.

Consider using the package [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.quiver.html) to add arrows for the vectors. Plot the arrows starting either at the origin or at the mean of your distribution. 

```python
# First, reproduce the plot from earlier. 
# Add arrows corresponding to the empirical eigenvalues. 
plt.scatter(X[:,0],X[:,1])
plt.quiver([mean[0]] * 2, [mean[1]] * 2, vals**-1 * vectors[:,0], vals**-1 * vectors[:,1], color=['r','r'], scale=4)
plt.show()
```

```python
# First, reproduce the plot from earlier. 
# Add arrows corresponding to the eigenvalues computed from the empirical covariance matrix.
plt.scatter(X[:,0],X[:,1])
plt.quiver([mean[0]] * 2, [mean[1]] * 2, vals**-1 * vectors[:,0], vals**-1 * vectors[:,1], color=['r','r'], scale=4)
plt.quiver([mean[0]] * 2, [mean[1]] * 2, empirical_vals**-1 * empirical_vectors[:,0],
           empirical_vals**-1 * empirical_vectors[:,1], color=['k','k'], scale=.04)
plt.show()
```
